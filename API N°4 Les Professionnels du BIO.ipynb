# 1 : Analyse des producteurs/vendeurs du 74 

```python
# Lien url de la request : 
link = 'https://opendata.agencebio.org/api/gouv/operateurs/?departements=74&nb=9000'

# Request du lien : 
r = requests.get(link)

# VÃ©rificiation de la bonne lecture de l'API : (200 = Good) :
r
```

```python
# Lecture de la request en json :
Bio = r.json()

# Visualiser un fichier json de faÃ§on plus structurÃ© :
Bio = pd.json_normalize(Bio['items'])

# Data Frame :
Bio.head(3)
```

```python
# Nettoyage de la DF :

# Def de rÃ©cupÃ©ration des noms ou les organismes dans les dico :
def RecuNoms(value) :
    ListeNoms = []

    for nbnoms in range(0, len(value)) :
        try :
            ListeNoms.append(value[nbnoms]['nom'])
        except :
            ListeNoms.append(value[nbnoms]['organisme'])

    return ListeNoms

# RÃ©cupÃ©ration des noms des catÃ©gories produits via def:
Bio['categories'] = Bio['categories'].apply(RecuNoms)

# RÃ©cupÃ©ration de l'url du site web :
Bio['siteWebs'] = Bio['siteWebs'].apply(lambda x : x[0]['url'] if len(x) >= 1 else 'Non CommuniquÃ©')

# RÃ©cupÃ©ration des informations de localisation :
# Si vide dans cette colonnes on supprime la colonne :
Bio['len'] = Bio['adressesOperateurs'].apply(lambda x : len(x))
Bio = Bio[Bio['len'] != 0]
del Bio['len']

Bio['Ville'] = Bio['adressesOperateurs'].apply(lambda x: x[0]['ville'])
Bio['CodePostal'] = Bio['adressesOperateurs'].apply(lambda x: x[0]['codePostal'])
Bio['Latitude'] = Bio['adressesOperateurs'].apply(lambda x: x[0]['lat'])
Bio['Longitude'] = Bio['adressesOperateurs'].apply(lambda x: x[0]['long'])

del Bio['adressesOperateurs']

# RÃ©cupÃ©ration des produits via une def :
Bio['productions'] = Bio['productions'].apply(RecuNoms)

# RÃ©cupÃ©ration des activitÃ©s via la def prÃ©cÃ©dente :
Bio['activites'] = Bio['activites'].apply(RecuNoms)

# RÃ©cupÃ©ration des organismes de certification :
Bio['certificats'] = Bio['certificats'].apply(RecuNoms)

# Suppression des colonnes inutiles :
Bio = Bio.drop(columns = ['siret','telephone', 'email', 'gerant', 'dateMaj', 'telephoneCommerciale', 'reseau'], axis = 1)
```

# 2 : Analyse des producteurs/vendeurs de France et DOM/TOM

## 2.1 : CrÃ©ation d'une liste des dÃ©partements

```python
# Lien url de la request : 
link = 'https://geo.api.gouv.fr/departements?'

# Request du lien : 
r = requests.get(link)

# Lecture de la request en json :
Departements = r.json()

# Visualiser un fichier json de faÃ§on plus structurÃ© :
Departements = pd.json_normalize(Departements)

# Data Frame :
Departements.head(3)
                  
# liste DÃ©partements :
ListDept = list(Departements['code'])
```

## 2.2 : RÃ©cupÃ©ration des donnÃ©es de toutes les rÃ©gions

```python
DFglobale = pd.DataFrame()

for dept in tqdm.tqdm(ListDept) :

    # Lien url de la request : 
    link = 'https://opendata.agencebio.org/api/gouv/operateurs/?departements={}&nb=9000'.format(dept)

    # Request du lien : 
    r = requests.get(link)

    # VÃ©rificiation de la bonne lecture de l'API : (200 = Good) :
    if r.status_code != 200 :
        print(dept)

    # Lecture de la request en json :
    Bio = r.json()

    # Visualiser un fichier json de faÃ§on plus structurÃ© :
    Bio = pd.json_normalize(Bio['items'])

    DFglobale = pd.concat([DFglobale, Bio])
```

```python
# Modification de la DF :
DFglobale.reset_index(drop = True, inplace = True)

# Nettoyage de la DF :

# Def de rÃ©cupÃ©ration des noms ou les organismes dans les dico :
def RecuNoms(value) :
    ListeNoms = []

    for nbnoms in range(0, len(value)) :
        try :
            ListeNoms.append(value[nbnoms]['nom'])
        except :
            ListeNoms.append(value[nbnoms]['organisme'])

    return ListeNoms

# RÃ©cupÃ©ration des noms des catÃ©gories produits via def:
DFglobale['categories'] = DFglobale['categories'].apply(RecuNoms)

# RÃ©cupÃ©ration de l'url du site web :
DFglobale['siteWebs'] = DFglobale['siteWebs'].apply(lambda x : x[0]['url'] if len(x) >= 1 else 'Non CommuniquÃ©')

# RÃ©cupÃ©ration des informations de localisation :
# Si vide dans cette colonnes on supprime la colonne :
DFglobale['len'] = DFglobale['adressesOperateurs'].apply(lambda x : len(x))
DFglobale = DFglobale[DFglobale['len'] != 0]
del DFglobale['len']

DFglobale['Ville'] = DFglobale['adressesOperateurs'].apply(lambda x: x[0]['ville'])
DFglobale['CodePostal'] = DFglobale['adressesOperateurs'].apply(lambda x: x[0]['codePostal'])
DFglobale['Latitude'] = DFglobale['adressesOperateurs'].apply(lambda x: x[0]['lat'])
DFglobale['Longitude'] = DFglobale['adressesOperateurs'].apply(lambda x: x[0]['long'])

del DFglobale['adressesOperateurs']

# RÃ©cupÃ©ration des produits via une def :
DFglobale['productions'] = DFglobale['productions'].apply(RecuNoms)

# RÃ©cupÃ©ration des activitÃ©s via la def prÃ©cÃ©dente :
DFglobale['activites'] = DFglobale['activites'].apply(RecuNoms)

# RÃ©cupÃ©ration des organismes de certification :
DFglobale['certificats'] = DFglobale['certificats'].apply(RecuNoms)

# Suppression des colonnes inutiles :
DFglobale = DFglobale.drop(columns = ['siret','telephone', 'email', 'gerant', 'dateMaj', 'telephoneCommerciale', 'reseau'], axis = 1)
```

```python
# RÃ©cupÃ©ration des donnÃ©es gÃ¨ographiques de la france :
url = 'https://github.com/gregoiredavid/france-geojson/raw/master/departements-version-simplifiee.geojson'
sf = gpd.read_file(url)
sf.head()
```
