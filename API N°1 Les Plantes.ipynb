# 3 : Analyse sur 100 pages et 20 plantes par page

## 3.1 : Récupération des noms des plantes

```python
# Liste des liens api de chaque plantes :
LiensPlantes = []

# Récupération des liens sur plusieurs pages :
for page in tqdm.tqdm(range (1, 101)) :
        
    # Lien url de chaque page :
    Token = '05if13gUgeosHfoyW3bZ3JO18chNyJvE1lNcS8JlYVE'
    link = 'https://trefle.io/api/v1/species?page={}&token={}'.format(page, Token)

    # Request du lien : 
    r = requests.get(link)
    
    # Vérification :
    if r.status_code != 200 :
        print(r)

    # Lecture de la request en json :
    data = r.json()

    # Récupértion de données de la page 1 : 
    df = pd.DataFrame(data['data'])

    # Garder les colonnes utiles :
    df = df[['links']]

    # Récupération des liens api :
    for i in range (0, len(df)) :
        LiensPlantes.append(df['links'][i]['self'])
```

# 3.2 : Utilisation de ces liens pour récupérer des informations de plantes 

```python
# Création d'une DF vide :
Especes = pd.DataFrame()


for lien in tqdm.tqdm(LiensPlantes) :

    # Lien url de la request :
    Token = '05if13gUgeosHfoyW3bZ3JO18chNyJvE1lNcS8JlYVE'
    link = 'https://trefle.io{}?token={}'.format(lien, Token)

    # Request du lien : 
    r = requests.get(link)

    # Lecture de la request en json :
    espece = r.json()

    # Visualiser un fichier json de façon plus structuré :
    espece = pd.json_normalize(espece)

    # Data Frame :
    Especes = pd.concat([espece, Especes])
```
